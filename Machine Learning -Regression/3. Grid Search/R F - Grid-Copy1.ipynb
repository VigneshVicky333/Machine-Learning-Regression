```python
#importing the Libraies
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Reading the Dataset
dataset = pd.read_csv('insurance_pre.csv')
```

```python
dataset
```

```python
dataset=pd.get_dummies(dataset,drop_first=True)
```

```python
dataset
```

```python
indep=dataset[['age', 'bmi', 'children','sex_male', 'smoker_yes']]
dep=dataset['charges']
```

```python
# Here we are going to remove the training set and test set, since it is having the cross validation

#split into training set and test
#from sklearn.model_selection import train_test_split
#X_train, X_test, y_train, y_test = train_test_split(indep, dep, test_size = 1/3, random_state = 0)
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
from sklearn.model_selection import GridSearchCV

#from sklearn.tree import DecisionTreeRegressor

param_grid = {'criterion':['squared_error', 'absolute_error', 'friedman_mse', 'poisson',],'max_features': ['sqrt','log2'],'n_estimators':[10,100]}

grid = GridSearchCV(RandomForestRegressor(), param_grid, refit = True, verbose= 3,n_jobs=-1)


# fitting the model for grid search, 
#here we are going to remove this x-train and y-train, since we are using the grid search

grid.fit(indep, dep)
```

```python
# print best parameter after tuning
#print(grid.best_params_)

re=grid.cv_results_
print("The R_score value for best parameter {}:".format(grid.best_params_))
```

```python
table=pd.DataFrame.from_dict(re)
```
