```python
import pandas as pd
```

```python
dataset=pd.read_csv("50_startups.csv")
```

```python
dataset
```

```python
dataset=pd.get_dummies(dataset,dtype=int,drop_first=True)
```

```python
dataset
```

```python
dataset.columns
```

```python
independent=dataset[['R&D Spend', 'Administration', 'Marketing Spend','State_Florida', 'State_New York']]   
```

```python
independent
```

```python
dependent=dataset[["Profit"]]
```

```python
dependent
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(independent, dependent, test_size=0.30,random_state=0)
```

```python
# As of now we have completed splitting the training set and test set, 
# Next we shall move to the model creation.
```

```python
from sklearn.tree import DecisionTreeRegressor
regressor=DecisionTreeRegressor(criterion='poisson', splitter='best')
regressor=regressor.fit(X_train,y_train)

# We should not forget that once the fit function is executed ---> Model is created nu,
# and the createdmodel will be stored or saved in that regressor variable.
```

```python
import matplotlib.pyplot as plt
from sklearn import tree
tree.plot_tree(regressor)
plt.show()


# Here pyplot as plt means for plotting the graph.
```

```python
# Next is evaluation matrix.

y_pred=regressor.predict(X_test)
```

```python
from sklearn.metrics import r2_score
r_score=r2_score(y_test,y_pred)
```
